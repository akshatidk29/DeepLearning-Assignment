```python
from tensorflow import keras
from tensorflow.keras import models,layers
print(f"TensorFlow version: {tf.__version__}")
print(f"GPU Available: {tf.config.list_physical_devices('GPU')}")
```

```python
num_classes = 10

model = models.Sequential([
    #first convolution layer
    layers.Conv2D(32,(3,3),padding='same',input_shape=(64,64,3)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    #second convolution layer
    layers.Conv2D(32,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    #first pooling layer
    layers.MaxPooling2D((2,2)),
    #third convolution layer
    layers.Conv2D(64,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    #fourth convolution layer
    layers.Conv2D(64,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    #second pooling layer
    layers.MaxPooling2D((2,2)),
    #fifth convolution layer
    layers.Conv2D(128,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    #sixth convolution layer
    layers.Conv2D(128,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    #third pooling layer
    layers.MaxPooling2D((2,2)),

    #now flattening and then applying the fcnn
    layers.Flatten(),
    layers.Dense(256,activation='relu'),
    layers.Dense(num_classes,activation='softmax')
])

model.summary()
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',  # Use 'categorical_crossentropy' if labels are one-hot encoded
    metrics=['accuracy']
)

print("Model compiled successfully!")
```

```python
train_dataset = tf.keras.utils.image_dataset_from_directory(
    "../tiny_imagenet/train",
    labels="inferred",          # labels from folder names
    label_mode="int",           # integer labels (use sparse loss)
    image_size=(64, 64),        # resize to match model input
    batch_size=32,
    shuffle=True
)

val_dataset = tf.keras.utils.image_dataset_from_directory(
    "../tiny_imagenet/val",
    labels="inferred",
    label_mode="int",
    image_size = (64,64),
    batch_size=32,
    shuffle=True
)
```

```python
# Train the model
EPOCHS = 100

tf.debugging.set_log_device_placement(True)

import time

start = time.time()

history = model.fit(
    train_dataset,              # Training data
    validation_data=val_dataset, # Validation data
    epochs=EPOCHS,               # Number of epochs
    verbose=1                    # Show progress
)

print(time.time() - start)
print("\nTraining complete!")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

train_dataset = tf.keras.utils.image_dataset_from_directory(
    "../tiny_imagenet/train",
    labels="inferred",
    label_mode="int",
    image_size=(64, 64),
    batch_size=32,
    shuffle=True
)

num_classes = len(train_dataset.class_names)

data_augmentation = tf.keras.Sequential([
    layers.RandomRotation(0.0833), 
])

def color_jitter(x):
    x = tf.image.random_brightness(x, 0.2)
    x = tf.image.random_contrast(x, 0.8, 1.2)
    x = tf.image.random_saturation(x, 0.8, 1.2)
    x = tf.image.random_hue(x, 0.1)
    return x

total_images = 0
for _ in train_dataset:
    total_images += 32  # batch_size

print(f"Total training images: {total_images}")

num_augment = int(total_images * 0.40)
augment_mask = np.zeros(total_images, dtype=np.float32)
augment_mask[:num_augment] = 1.0
np.random.shuffle(augment_mask)  # Shuffle once

print(f"Images to augment: {num_augment} ({num_augment/total_images*100:.1f}%)")

augment_mask = tf.constant(augment_mask, dtype=tf.float32)

train_dataset_unbatched = train_dataset.unbatch()
def add_index(idx, sample):
    image, label = sample
    return idx, image, label

train_dataset_indexed = train_dataset_unbatched.enumerate().map(
    lambda idx, sample: (idx, sample[0], sample[1]),
    num_parallel_calls=tf.data.AUTOTUNE
)


def augment_by_index(idx, x, y):
    x = tf.cast(x, tf.float32) / 255.0
    should_augment = augment_mask[idx]
    augmented = data_augmentation(tf.expand_dims(x, 0), training=True)[0]
    augmented = color_jitter(augmented)
    mask = tf.reshape(should_augment, [1, 1, 1])
    x = mask * augmented + (1 - mask) * x
    
    return x, y

train_dataset_processed = train_dataset_indexed.map(
    augment_by_index,
    num_parallel_calls=tf.data.AUTOTUNE
)

train_dataset_processed = train_dataset_processed.batch(32)
train_dataset_processed = train_dataset_processed.prefetch(tf.data.AUTOTUNE)

print("Dataset prepared with fixed 20% augmentation!")

test_dataset = tf.keras.utils.image_dataset_from_directory(
    "../tiny_imagenet/val",
    labels="inferred",
    label_mode="int",
    image_size=(64, 64),
    batch_size=32,
    shuffle=False
)

test_dataset = test_dataset.map(
    lambda x, y: (tf.cast(x, tf.float32) / 255.0, y),
    num_parallel_calls=tf.data.AUTOTUNE
)

test_dataset = test_dataset.prefetch(tf.data.AUTOTUNE)

model = models.Sequential([
    layers.Input(shape=(64, 64, 3)),

    layers.Conv2D(32,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),

    layers.Conv2D(32,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),

    layers.Conv2D(64,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),

    layers.Conv2D(128,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(256,activation='relu'),
    layers.Dense(num_classes,activation='softmax')
])


model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

history_after_augumentation = model.fit(
    train_dataset_processed,
    epochs=EPOCHS
)
test_loss, test_accuracy = model.evaluate(test_dataset)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```
