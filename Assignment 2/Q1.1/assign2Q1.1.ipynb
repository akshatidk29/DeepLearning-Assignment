```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class CNN(nn.Module):

    def __init__(self):
        super(CNN, self).__init__()

#3 Convolution Layers (3×3)
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=64, kernel_size=3)
        self.conv2 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3)
        self.conv3 = nn.Conv2d(in_channels=128, out_channels=3, kernel_size=3)

# 2 MaxPooling Layers (2×2 stride 2)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

 # Fully Connected Layer
        self.fc = nn.Linear(14 * 14 * 3, 10)

    def forward(self, x):

        print("Input:", x.shape)

        x = F.relu(self.conv1(x))     
        print("After Conv1 + ReLU:", x.shape)

        x = F.relu(self.conv2(x))    
        print("After Conv2 + ReLU:", x.shape)

        x = self.pool(x)
        print("After Pool1:", x.shape)

        x = F.relu(self.conv3(x))    
        print("After Conv3 + ReLU:", x.shape)

        x = self.pool(x)
        print("After Pool2:", x.shape)

        x = torch.flatten(x, start_dim=1)
        print("After Flatten:", x.shape)

        x = self.fc(x)
        print("After FC:", x.shape)

        return x


# test input 64 × 64 × 3


model = CNN()
dummy_input = torch.randn(1, 3, 64, 64)

output = model(dummy_input)
```
