```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset
```

```python
train_path = r"dataset/tiny-imagenet/train"
```

```python
class CNN(nn.Module):

    def __init__(self):
        super(CNN, self).__init__()

        self.conv1 = nn.Conv2d(3, 64, 3)
        self.conv2 = nn.Conv2d(64, 128, 3)
        self.conv3 = nn.Conv2d(128 ,256, 3, )

        self.pool = nn.MaxPool2d(2, 2)

        self.fc = nn.Linear(256 * 14 * 14, 10)

    def forward(self, x):

        print("Input:", x.shape)

        x = F.relu(self.conv1(x))
        print("After Conv1 + ReLU:", x.shape)

        x = F.relu(self.conv2(x))
        print("After Conv2 + ReLU:", x.shape)

        x = self.pool(x)
        print("After Pool1:", x.shape)

        x = F.relu(self.conv3(x))
        print("After Conv3 + ReLU:", x.shape)

        x = self.pool(x)
        print("After Pool2:", x.shape)

        x = torch.flatten(x, start_dim=1)
        print("After Flatten:", x.shape)

        x = self.fc(x)
        print("After FC:", x.shape)

        return x
```

```python
# Transforms

transform = transforms.Compose([
    transforms.Resize((64, 64)),  # keeps stable dimensions
    transforms.ToTensor()
])
```

```python
# Load full dataset

full_dataset = datasets.ImageFolder(
    root=train_path,
    transform=transform
)
```

```python
# Filter only 10 Classes

selected_classes = list(range(10))  # labels from 0 – 9

indices = [
    i for i, (_, label) in enumerate(full_dataset)
    if label in selected_classes
]

small_dataset = Subset(full_dataset, indices)


print("Total images used:", len(small_dataset))
```

```python
# Data Loader

train_loader = DataLoader(
    small_dataset,
    batch_size=1,      
    shuffle=True
)
```

```python
model = CNN()
images, labels = next(iter(train_loader))
output = model(images)
```
